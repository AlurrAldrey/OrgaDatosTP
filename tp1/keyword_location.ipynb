```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('train.csv')
train.head()
```

```python
disaster_tweets = train.loc[train['target'] == 1]
non_disaster_tweets = train.loc[train['target'] == 0]
```

```python
disaster_tweets.keyword.describe()
```

```python
non_disaster_tweets.keyword.describe()
```

```python
disaster_tweets.keyword.isna().sum()
```

```python
 non_disaster_tweets.keyword.isna().sum()
```

```python
# Keyword mas frecuentes
plt.figure(figsize=(9,6))
sns.countplot(y=train.keyword, order = train.keyword.value_counts().iloc[:15].index)
plt.title('Top 15 keywords mas frecuentes')
plt.show()
```

```python
#Keywords mas frecuentes segun su target
keyword_disaster = disaster_tweets.keyword.value_counts().head(15)
keyword_non_disaster = non_disaster_tweets.keyword.value_counts().head(15)

plt.figure(figsize=(13,5))
plt.subplot(121)
sns.barplot(keyword_disaster,keyword_disaster.index, color='c')
plt.title('Top 15 keywords para disaster tweets')
plt.subplot(122)
sns.barplot(keyword_non_disaster, keyword_non_disaster.index, color='y')
plt.title('Top 15 keywords para non-disaster tweets')
plt.show()
```

```python
keyword_disaster
```

```python
#Se puede ver que las keywords mas frecuentes de los disaster tweets y de los non-disaster tweets son diferentes.

#Si se suman los terminos 'suicide bombing', 'suicide bomb' y 'suicide bomber', que remiten a lo mismo, se alcanza una frecuencia de 94 y se posiciona como la keyword con mas apariciones dentro de los disaster tweets.

#Las keywords de los non-disaster tweets suelen ser terminos mas sensacionalistas que las de los disaster tweets. Un ejemplo de eso son: 'armageddon', 'fear', 'panic' y 'screaming'. 
```
